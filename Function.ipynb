```python
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
```

```python
def funct (documents,directorie):
    while True:
        #print('- p найти сотрудника по номеру документа')
        
        user_input = input('Введите команду:')
        
    
        if user_input == 'p':
            doc_input = input('Введите номер документа:')
            for document in documents:
                flag = 1
                number_1 = document['number']
                name = document['name']
                if doc_input == number_1:
                    print(f'Владелец документа: {name}')
                    flag = 0
            if flag :
                print('Владелец документа не найден')
        elif user_input=='s':
            doc_input = input('Введите номер документа:')
            for k in directories:
                pl=1
                if doc_input in directories[k]:
                    print(f'Полка: {k}')    
                    pl=0
            if pl:
                print('Документ не найден в базе')
        elif user_input=='j':
            for k in directories:
                if documents[0]['number'] in directories[k]: 
                    print ('Номер документа:', documents[0]['number'], 'type:',documents[0]['type'],'владелец:', documents[0]['name'],f'Полка: {k}')
            for k in directories:
                if documents[1]['number'] in directories[k]:         
                    print ('Номер документа:', documents[1]['number'], 'type:',documents[1]['type'],'владелец:', documents[1]['name'],f'Полка: {k}')
            for k in directories:
                if documents[2]['number'] in directories[k]:         
                    print ('Номер документа:', documents[2]['number'], 'type:',documents[2]['type'],'владелец:', documents[2]['name'],f'Полка: {k}')
        elif user_input=='as':
                a=dict.keys(directories)
                dir_input = input('Введите номер полки:')            
                if dir_input in directories:
                    print('Полка уже существует', 'Текущий перечень полок:', list(a))
                else:
                    directories[dir_input]=[]
                    print('Полка добавлена', 'Текущий перечень полок:', list(a))
        elif user_input=='ds':
            a1=dict.keys(directories)  
            dir1_input = input('Введите номер полки:')
            if dir1_input  in directories:
                if len(directories[dir1_input])>0:
                    print('На полке содержатся документы, сначала удалите их', 'Текущий перечень полок:', list(a1))                               
                else:
                    del(directories[dir1_input])
                    print('Полка удалена', 'Текущий перечень полок:', list(a1))    
            else:
                print ('Такой полки не существует. Текущий перечень полок:',list(a1))                   
       
        elif user_input == 'q':         
                print('Выход из программы')
                break
        else:
            print('Введена неверная команда')
        
funct (documents,directories)
```
